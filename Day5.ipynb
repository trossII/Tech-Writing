Yesterday I spent the day learning about Natural Language Processing. What I found the most interesting and will probably be using a ton in the future is the TF-IDF score. The TF stands for Term Frequency and the IDF is Inverse Document Frequency. So when you mulipy these two values together you get the amount of times a word is used in its own specific text multiplied by the log of it occurs regularly in the other texts as well. Therefore scoring the word higher the more unique it is to specific texts so that you can better model differences between your documents.

Normally you would clean the data first but I am still getting used to used to using the TfidfVectorizer() so below is a very basic example.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
```

```python
# read in data
df_messages = pd.read_csv('data/spam.csv', usecols=[0,1])

# convert string labels to 1 or 0 
le = LabelEncoder()
df_messages['target'] = le.fit_transform(df_messages['v1'])

X = df_messages['v2']
y = df_messages['target']
```

```python

vectorizer = TfidfVectorizer()
tf_idf = vectorizer.fit_transform(X)
nlpskl = pd.DataFrame(tf_idf.toarray(), columns=vectorizer.get_feature_names())
nlpskl.head()
```
